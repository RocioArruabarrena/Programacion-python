```python
#Ejercicio N°1
# Se quiere saber cuantas localidades pertenecen a cada continente.
# > La respuesta debe ser un diccionario donde los nombres de los continentes sean las claves y el valor sea la cantidad 
# de ciudades que corresponden a cada uno.

#sumar los elementos y sumar continente 

def leer_datos(nombre_archivo, cabecera = []):
    ''' Devuelve una lista de diccionarios con la información del datos del archivo especificado '''
    import csv
    
    datos_dict = []
    try:
        file = open(nombre_archivo, 'rt', encoding='utf8')
        filas = csv.reader(file)
        if len(cabecera) == 0:
            cabecera = next(filas)
        for fila in filas:
            item =  dict(zip(cabecera,fila))
            datos_dict.append(item)
        file.close()
    except Exception as e:
        datos_dict.append({"error": f'A sucedido un Error {e}'})
    finally:
        return datos_dict

misDatos = leer_datos('data.csv')

#cuantas veces se repite el pais
resp = {}
for item in misDatos:
    pais = item['country']
    if not pais in resp:
        resp[pais] = 0
    resp[pais] += 1
    
print(resp)

#{ }

paises_continentes = {
    'AM': 'América',
    'EU': 'Europa',
    'AF': 'África',
    'AS': 'Asia',
    'OC': 'Oceanía'
}



paises_continentes.sort()
print(paises_continentes)















```

```python
# Mapa de países a continentes
paises_a_continentes = {
    "Argentina": "América del Sur",
    "Brasil": "América del Sur",
    "Estados Unidos": "América del Norte",
    "Canadá": "América del Norte",
    "Russia": "Europa",
    "Francia": "Europa",
    "China": "Asia",
    "Japón": "Asia",
    # Agrega aquí más países y sus continentes
}

def contar_ciudades_por_continente(misDatos, paises_a_continentes):
    # Diccionario para contar ciudades por continente
    continentes = {}

    # Iteramos sobre los datos
    for item in misDatos:
        pais = item.get('country')
        
        # Verificamos si el país está en el diccionario de paises_a_continentes
        if pais in paises_a_continentes:
            continente = paises_a_continentes[pais]
            
            # Si el continente no está en el diccionario continentes, lo inicializamos en 0
            if continente not in continentes:
                continentes[continente] = 0
            
            # Aumentamos el contador de ciudades para ese continente
            continentes[continente] += 1
    
    return continentes

# Ejecución del código
misDatos = leer_datos('data.csv')
resultado = contar_ciudades_por_continente(misDatos, paises_a_continentes)

# Imprimimos el resultado
print(resultado)
```

```python
### Ejercicio N°3
# Se pide obtener la cantidad y los nombres de las localidades que pertenecen a un pais ingresado por el usuario
# > Se debera devolver en una funcion que tenga por parametro el "nombre_pais" y si devuelva una tupla 
# donde su primer valor sea la cantidad de localidades y su segundo valor sea una lista ordenada ascendente de los nombres de la ciudades.+




def leer_datos(nombre_archivo, cabecera = []):
    ''' Devuelve una lista de diccionarios con la información del datos del archivo especificado '''
    import csv
    
    datos_dict = []
    try:
        file = open(nombre_archivo, 'rt', encoding='utf8')
        filas = csv.reader(file)
        if len(cabecera) == 0:
            cabecera = next(filas)
        for fila in filas:
            item =  dict(zip(cabecera,fila))
            datos_dict.append(item)
        file.close()
    except Exception as e:
        datos_dict.append({"error": f'A sucedido un Error {e}'})
    finally:
        return datos_dict

misDatos = leer_datos('data.csv')




def obtener_localidades(nombre_pais):
    if nombre_pais in misDatos:
        citys = misDatos [nombre_pais]
        citys_ordenadas = sorted(citys)
        return (len(citys_ordenadas), citys_ordenadas)
    else:
        return (0, []) 

#Ejemplo 
counntry_ingresado = input("Ingresa el nombre del país: ")
cantidad_localidades, citys = obtener_localidades(counntry_ingresado)

print(f"Cantidad de localidades en {counntry_ingresado}: {cantidad_localidades}")
print("Localidades:", citys)
```
