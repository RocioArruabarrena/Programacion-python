```python
### Ejercicio N°1
# Se quiere sabe cuantos partidos fueron empate y cuales fueron los resultados ordenados de mayor a menor.
# > La respuesta debe ser un una diccionario donde la primra clave contenga la cantidad de empates y 
# la segunda clave sean los resultados en una lista de tuplas, no repetidas, por ejemplo 
# { "empates": 15, "resultados": [(1-1),(3-3). ...]}



def leer_datos(nombre_archivo, cabecera = []):
    ''' Devuelve una lista de diccionarios con la información del datos del archivo especificado '''
    import csv
    
    datos_dict = []
    try:
        file = open(nombre_archivo, 'rt', encoding='utf8')
        filas = csv.reader(file)
        if len(cabecera) == 0:
            cabecera = next(filas)
        for fila in filas:
            item =  dict(zip(cabecera,fila))
            datos_dict.append(item)
        file.close()
    except Exception as e:
        datos_dict.append({"error": f'A sucedido un Error {e}'})
    finally:
        return datos_dict

misDatos = leer_datos('partidos.csv')

empates = 0
resultado = []


for item in misDatos:
    resultado_str = item['resultado'].split('-')
    
    
    golesA, golesB = map(int, resultado_str)
    
    resultado.append((golesA, golesB))
    
   
    if golesA == golesB:
        empates += 1

resultado_unico = sorted(set(resultado), reverse=True)

resultado_final = {
    "empates": empates,
    "resultado": resultado_unico
}

print(resultado_final)


```
