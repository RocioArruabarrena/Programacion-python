```python
### Ejercicio N°3
# Se pide obtener la cantidad y los nombres de las localidades que pertenecen a un pais ingresado por el usuario
# > Se debera devolver en una funcion que tenga por parametro el "nombre_pais" y si devuelva una tupla 
# donde su primer valor sea la cantidad de localidades y su segundo valor sea una lista ordenada ascendente de los nombres de la ciudades.+




def leer_datos(nombre_archivo, cabecera = []):
    ''' Devuelve una lista de diccionarios con la información del datos del archivo especificado '''
    import csv
    
    datos_dict = []
    try:
        file = open(nombre_archivo, 'rt', encoding='utf8')
        filas = csv.reader(file)
        if len(cabecera) == 0:
            cabecera = next(filas)
        for fila in filas:
            item =  dict(zip(cabecera,fila))
            datos_dict.append(item)
        file.close()
    except Exception as e:
        datos_dict.append({"error": f'A sucedido un Error {e}'})
    finally:
        return datos_dict

misDatos = leer_datos('data.csv')



def obtener_localidades(nombre_pais):
    
    localidades = []
    
    
    for item in misDatos:
        if item['country'] == nombre_pais:
            localidades.append(item['city'])  
    
    localidades_ordenadas = sorted(localidades)
    
    return len(localidades_ordenadas), localidades_ordenadas

counntry_ingresado = input("Ingresa el nombre del país: ")
cantidad_localidades, localidades = obtener_localidades(counntry_ingresado)

print(f"cantidad de localidades: {counntry_ingresado}: {cantidad_localidades}")
print("localidades:", localidades)

```
